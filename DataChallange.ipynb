```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 



Data=pd.read_csv('employee_retention_data.csv',sep=',')

Data['Has-quit']=~Data['quit_date'].isnull()
```

```python
Company_id_Retention_rate=Data.groupby(['company_id'])['Has-quit'].value_counts(normalize=True).rename('percentage').mul(100).reset_index()
```

```python
Company_id_Retention_rate
```

```python
Company_id_Retention_rate=Company_id_Retention_rate[Company_id_Retention_rate['Has-quit']==True]
```

```python
objects = Company_id_Retention_rate['company_id']
y_pos = np.arange(len(objects))
performance = Company_id_Retention_rate['percentage']
 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Employee retention percentage')
plt.xlabel('Company ID')
plt.title('Employee retention rate based on company ID')
 
plt.show()
```

```python
def make_dummies(df, variables):
    
    dummies = pd.get_dummies(df[variables], prefix = variables)
    df = pd.concat([df, dummies], axis = 1)
    df = df.drop(variables, axis = 1)
    return df
```
